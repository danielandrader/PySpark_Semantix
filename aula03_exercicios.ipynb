## Exercícios - RDD 

```python
#1. Ler com RDD os arquivos localmente do diretório “/opt/spark/logs/” ("file:///opt/spark/logs/")
```

```python
rdd = sc.textFile("file:///opt/spark/logs/*")
```

```python
#2. Com uso de RDD faça as seguintes operações

#a) Contar a quantidade de linhas
```

```python
rdd.count()
```

```python
#b) Visualizar a primeira linha
```

```python
rdd.first()
```

```python
# c) Visualizar todas as linhas
```

```python
rdd.collect()
```

```python
# d) Contar a quantidade de palavras
```

```python
palavras = rdd.flatMap(lambda x: x.split(" "))
```

```python
palavras.collect()
```

```python
palavras.count()
```

```python
palavras_map = rdd.map(lambda x: x.split(" "))
```

```python
palavras_map.collect()
```

```python
# e) Converter todas as palavras em minúsculas
```

```python
minusculas = palavras.map(lambda linha: linha.lower())
```

```python
minusculas.collect()
```

```python
# f) Remover as palavras de tamanho menor que 2
```

```python
palavras_2 = palavras.filter(lambda palavras: len(palavras) > 2)
```

```python
palavras_2.collect()
```

```python
# g) Atribuir o valor de 1 para cada palavra
```

```python
palavras_chaveValor = palavras.map(lambda palavras: (palavras, 1))
```

```python
palavras_chaveValor.collect()
```

```python
# h) Contar as palavras com o mesmo nome
```

```python
palavras_chaveValor_reduce = palavras_chaveValor.reduceByKey(lambda key1, key2: key1 + key2)
```

```python
palavras_chaveValor_reduce.take(10)
```

```python
palavras_chaveValor_reduce.collect()
```

```python
# i) Visualizar em ordem alfabética
```

```python
palavras_chaveValor_reduce.order = palavras_chaveValor_reduce.sortBy(lambda palavras_chaveValor_reduce: palavras_chaveValor_reduce[0], True)
```

```python
palavras_chaveValor_reduce.order.collect()
```

```python
# j) Visualizar em ordem decrescente a quantidade de palavras
```

```python
palavras_chaveValor_reduce.order_desc = palavras_chaveValor_reduce.sortBy(lambda palavras_chaveValor_reduce: palavras_chaveValor_reduce[1], False)
```

```python
palavras_chaveValor_reduce.order_desc.collect()
```

```python
# k) Remover as palavras, com a quantidade de palavras > 1
```

```python
palavras_chaveValor_reduce_1 = palavras_chaveValor_reduce.filter(lambda palavras_chaveValor_reduce: palavras_chaveValor_reduce[1]==1)
```

```python
palavras_chaveValor_reduce_1.collect()
```

```python
# l) Salvar o RDD no diretorio do HDFS /user/daniel/logs_count_word
```

```python
palavras_chaveValor_reduce_1.getNumPartitions()
```

```python
palavras_chaveValor_reduce_1.saveAsTextFile("/user/daniel/logs_count_word")
```

```python
!hdfs dfs -ls /user/daniel/logs_count_word
```
