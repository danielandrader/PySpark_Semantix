## WithColumn 2/2 

```python
# 1. Criar um dataframe para ler o arquivo no HDFS /user/<nome/data/juros_selic/juros_selic
```

```python
df_juros = spark.read.csv("/user/daniel/data/exercises-data/juros_selic/juros_selic", sep=";", header="True")
```

```python
df_juros.show(5)
```

```python
# 2. Agrupar todas as datas pelo ano em ordem decrescente e salvar a quantidade de meses ocorridos, o valor médio, mínimo e máximo do campo valor com a seguinte estrutura:
```

```python
df_juros_ano = df_juros.withColumn("ano", from_unixtime(unix_timestamp(col("data"), "dd/MM/yyyy"),"yyyy"))
df_juros_ano_mes = df_juros_ano.withColumn("mes", from_unixtime(unix_timestamp(col("data"), "dd/MM/yyyy"),"MM"))
```

```python
df_juros_ano_mes_valor = df_juros_ano_mes.withColumn("valor", regexp_replace(col("valor"), "\.", ""))
df_juros_ano_mes_valor = df_juros_ano_mes_valor.withColumn("valor", regexp_replace(col("valor"), "\,", "."))
```

```python
df_juros_ano_mes_valor = df_juros_ano_mes_valor.withColumn("valor", col("valor").cast(FloatType()))
```

```python
df_juros_ano_mes_valor.show(5)
```

```python
df_juros_ano_mes_valor.printSchema()
```

```python
from pyspark.sql.functions import count, avg, min, max, desc, format_number
```

```python
df_juros_ano_mes_valor.groupBy("ano").agg(count("mes").alias("Meses"),
                                          avg("valor").cast("decimal(38,2)").alias("Valor Médio"),
                                          min("valor").cast("decimal(38,2)").alias("Valor Mínimo"),
                                          max("valor").cast("decimal(38,2)").alias("Valor Máximo")
                                         ).sort(desc("ano")).show()
```

```python
df_juros_ano_mes_valor.groupBy("ano").agg(count("mes").alias("Meses"),
                                          format_number(avg("valor"),2).alias("Valor Médio"),
                                          format_number(min("valor"),2).alias("Valor Mínimo"),
                                          format_number(max("valor"),2).alias("Valor Máximo")
                                         ).sort(desc("ano")).show()
```

```python
# 3. Salvar no hdfs:///user/<nome>/relatorioAnual com compressão zlib e formato orc
```

```python
df_juros_ano_mes_valor.write.orc("/user/daniel/relatorioAnual", compression="zlib")
```
